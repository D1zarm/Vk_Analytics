## Для начала обучим модель линейной регрессии

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse, r2_score, mean_absolute_error as mae
import pandas as pd
from joblib import dump
```

```python
df = pd.read_csv('final_dataset.csv').set_index('user_id')
```

```python
len(df)
```

```python
df.head()
```

```python
# Transform Y_test['bdate'] and Y_pred from timestamp to year
def timestamp_to_year(Y_pred, Y_test):
    for idx, y in enumerate(Y_pred):
        Y_pred[idx] = datetime.fromtimestamp(y).year
        Y_test[idx] = datetime.fromtimestamp(Y_test[idx]).year
    return Y_pred, Y_test
```

```python
def MAE(Y_test, Y_pred):
    er=[]
    for i, test in enumerate(Y_test):
        a = abs(Y_pred[i] - test)
        er.append(a)

    print('MAE_test = {}'.format(sum(er)/len(er)))
    print('Top 10 outliers: {}'.format(sorted(er, reverse=True)[:10]))
    
    return er
```

```python
def train(df, regr, train_perc=20):
    
    n_test = int(len(df)*train_perc/100)

    Y_test = df['bdate'][-n_test:].to_numpy()
    Y_train = df['bdate'][:-n_test]

    print('Train size: {}'.format(len(Y_train)))
    print('Test size: {}'.format(len(Y_test)))
    
    df.drop(columns=['bdate', 'text_likes'], inplace = True)

    X_test = df[-n_test:]
    X_train = df[:-n_test]

    regr.fit(X_train, Y_train)
    print('R2_train = {}'.format('%.3f'%regr.score(X_train, Y_train)))
    print('R2_test = {}'.format('%.3f'%regr.score(X_test, Y_test)))

    Y_pred = regr.predict(X_test)

    Y_pred, Y_test = timestamp_to_year(Y_pred, Y_test)
    
    error = MAE(Y_test, Y_pred)
    print('MSE = {}'.format('%.3f'%mse(Y_test, Y_pred, squared= False)))
    
    return X_test, Y_test, Y_pred, error, regr
```
